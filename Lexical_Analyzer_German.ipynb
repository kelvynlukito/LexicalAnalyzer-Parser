```python
def lexicalAnalyzer():
    idx_char = 0
    state = 'q1'
    current_token = ''
    while state != "ACCEPT":
        current_char = text[idx_char]
        current_token += current_char
        state = transition_list[(state, current_char)]
        if (state == "q25"):
            print("Current Token: ", current_token, " valid")
            current_token = ""
        if (state == "q52"):
            print("Current Token: ", current_token, " valid")
            current_token = ""
        if (state == "q58"):
            print("Current Token: ", current_token, " valid")
            current_token = ""
        if state == "error":
            #document.getElementById("result-lexical").innerHTML = "error";
            print("error")
            break;
        idx_char = idx_char + 1
    # conclusion
    if state == "ACCEPT":
        print("Semua Token di Input: ", inputValue, " valid")

  
```
